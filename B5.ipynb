```python
import os
import glob
import argparse
import xml.etree.ElementTree as ET

# def parse_args():
#     parser = argparse.ArgumentParser(description='Convert YOLO annotations to VOC format')
#     parser.add_argument('input_folder', help='Path to input folder')
#     parser.add_argument('output_folder', help='Path to output folder')
#     return parser.parse_args()

def main():
    # args = parse_args()
    # os.makedirs(args.output_folder, exist_ok=True)
    for file_path in glob.glob( '*.txt'):
        with open(file_path) as f:
            lines = f.readlines()
        xml_path =  os.path.splitext(os.path.basename(file_path))[0] + '.xml'
        root = ET.Element('annotation')
        folder = ET.SubElement(root, 'folder')
        folder.text = 'images'
        filename = ET.SubElement(root, 'filename')
        filename.text = os.path.splitext(os.path.basename(file_path))[0] + '.jpg'
        size = ET.SubElement(root, 'size')
        width = ET.SubElement(size, 'width')
        height = ET.SubElement(size, 'height')
        depth = ET.SubElement(size, 'depth')
        width.text = '640'
        height.text = '480'
        depth.text = '3'
        for line in lines:
            parts = line.strip().split()
            obj = ET.SubElement(root, 'object')
            name = ET.SubElement(obj, 'name')
            name.text = parts[0]
            bndbox = ET.SubElement(obj, 'bndbox')
            xmin = ET.SubElement(bndbox, 'xmin')
            ymin = ET.SubElement(bndbox, 'ymin')
            xmax = ET.SubElement(bndbox, 'xmax')
            ymax = ET.SubElement(bndbox, 'ymax')
            xmin.text = str(int(float(parts[1]) * 640))
            ymin.text = str(int(float(parts[2]) * 480))
            xmax.text = str(int(float(parts[3]) * 640))
            ymax.text = str(int(float(parts[4]) * 480))
        tree = ET.ElementTree(root)
        tree.write(xml_path)

if __name__ == '__main__':
    main()
```
